```python
import json
import os
import shutil

# Đường dẫn
image_root = '/content/drive/MyDrive/OTU-2D-Dataset-main/OTU-2D-Dataset-main'
mask_root = '/content/drive/MyDrive/OTU-2D-Dataset-main/OTU-2D-Dataset-main'
json_root='/content/drive/MyDrive/OTU-2D-Dataset-main/OTU-2D-Dataset-main/OTU_2D_annotation.json'
```

```python
import os

json_folder = '/content/drive/MyDrive/OTU-2D-Dataset-main/OTU-2D-Dataset-main'
for file in os.listdir(json_folder):
    print(file)
```

```python
# Load file JSON
with open(json_root) as f:
    split_data = json.load(f)
```

```python
from collections import Counter
print(Counter([item['spalit'] for item in split_data]))
```

```python
# Tạo dict chứa dữ liệu chia theo split
split_data_dict = {'train': [], 'validation': [], 'test': []}

for item in split_data:
    split = item['split']
    if split == 'val':
        split = 'validation'

    image_path = item['file_path_img']
    mask_path = item['file_path_ann']
    if split in split_data_dict:
        split_data_dict[split].append((image_path, mask_path))
```

```python
output_root = '/content/drive/MyDrive/OTU-2D-Dataset-main/OTU-2D-Dataset-main/dataset_split'
```

```python
# Chia ảnh và mask vào các thư mục
for split in ['train', 'validation', 'test']:
    for img_rel_path, mask_rel_path in split_data_dict[split]:
        img_src = os.path.join(image_root, img_rel_path)
        mask_src = os.path.join(image_root, mask_rel_path)

        img_dst_dir = os.path.join(output_root, split, 'images')
        mask_dst_dir = os.path.join(output_root, split, 'masks')

        os.makedirs(img_dst_dir, exist_ok=True)
        os.makedirs(mask_dst_dir, exist_ok=True)

        shutil.copy(img_src, img_dst_dir)
        shutil.copy(mask_src, mask_dst_dir)

print(" Hoàn tất chia dữ liệu thành train, val, test!")
```
